```python
## replace all 60 duration with 50
```

```python
df=pd.read_csv('data/dirtydata.csv')
for x in df.index:
    print('index : ',x,'\nCurrent value = ',df.loc[x,'Duration'])
    if df.loc[x,"Duration"] == 60:
        df.loc[x,"Duration"] = 50
        print('Updated value : ',df.loc[x,"Duration"])
    
    print('-----------------------')
```

```python
df
```

```python
## print values greater than 60 , delete the row
df=pd.read_csv('data/dirtydata.csv')
for x in df.index:
    if df.loc[x,'Duration'] > 60:
        df.drop(x,inplace=True)

# to reset the index
df.reset_index(drop=True,inplace=True)
df
```
