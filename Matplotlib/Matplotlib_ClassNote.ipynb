```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel("Temperature")
plt.title('Weather info',fontsize=20)

plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='best',shadow=True)
plt.grid()
```

```python
# Bar chart  - x axis= categorical var


# Simple bar chart showing revenues of major US tech companies

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
```

```python
plt.bar(company,revenue)
plt.xlabel('Company')
plt.ylabel("Revenue")
plt.title('Revenue model for Top Companies')
```

```python
plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label="Profit")
plt.xlabel('Company')
plt.ylabel("Profit")
plt.title('Revenue and profit model for Top Companies')
plt.legend()
```

```python
## To make them side by side

xpos=np.arange(len(company))
xpos
```

```python
xpos=np.arange(len(company))

plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label="Profit")
plt.xlabel('Company')
plt.ylabel("Profit")
plt.title('Revenue and profit model for Top Companies')
plt.legend()
```

```python
xpos=np.arange(len(company))
plt.xticks(xpos,company)

plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label="Profit")
plt.xlabel('Company')
plt.ylabel("Profit")
plt.title('Revenue and profit model for Top Companies')
plt.legend()
```

```python
xpos=np.arange(len(company))
plt.xticks(xpos,company)

plt.bar(xpos+0.4,revenue,label='Revenue',width=0.4)
plt.bar(xpos,profit,label="Profit",width=0.4)
plt.xlabel('Company')
plt.ylabel("Profit")
plt.title('Revenue and profit model for Top Companies')
plt.legend()
```

```python
# for horrizontal plotting

plt.barh(company,profit)
```

```python
# Histogram           - x axis= numerical var , y axis=  frequency of that var


blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
# plt.show()
```

```python
plt.hist(blood_sugar,rwidth=0.8,color='green',bins=3)      # by default bins = 10
```

```python
plt.hist(blood_sugar,rwidth=0.8,color='black',bins=[80,100,125,150])
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women],label=['men','women'],bins=3,color=['grey','orange'])
plt.xlabel('Sugar level')
plt.ylabel('No of patients')

plt.legend()
```

```python
# Box Plot



#creating data set
np.random.seed(10)    # to get fixed random values for everyone
data = np.random.normal(100,20,200)  
# data

plt.figure(figsize=(10,5)) # can use anywhere

plt.boxplot(data)
plt.show()
```

```python
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]

fig = plt.figure(figsize =(10, 7))


# Creating plot
plt.boxplot(data)

# show plot
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(data,vert=False,patch_artist=True)
plt.show()
```

```python
# Pie chart

# Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.figure(figsize=(10,7))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%0.2f%%',shadow=True,explode=[0,0.1,0,0,0],startangle=90)
                                                   #values                                        #rotation
plt.legend(shadow=True)
# plt.show()

#saving
plt.savefig('sample_piechart.jpg')    # to show this , comment plt.show()
```
