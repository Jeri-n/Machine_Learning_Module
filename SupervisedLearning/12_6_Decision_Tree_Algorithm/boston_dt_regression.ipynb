```python
# importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore')

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
boston = load_boston()

df = pd.DataFrame(boston.data,columns=boston.feature_names)
df['Price'] = boston.target
df.head()
```

```python
df.isna().sum()
```

```python
X= df.drop(['Price'],axis=1)
y = df.Price

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train,y_train)

train_score_dt = dt_reg.score(X_train, y_train)
test_score_dt = dt_reg.score(X_test, y_test)

print(f"The Train score for  model is {train_score_dt}")
print(f"The Test score for  model is {test_score_dt}")

y_pred = dt_reg.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('MSE:',mse)
print('RMSE:',mse**(1/2))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(25,15))
plot_tree(dt_reg,filled=True)
plt.show()
```

```python
dt_reg = DecisionTreeRegressor(max_depth=7)
dt_reg.fit(X_train,y_train)

train_score_dt = dt_reg.score(X_train, y_train)
test_score_dt = dt_reg.score(X_test, y_test)

print(f"The Train score for  model is {train_score_dt}")
print(f"The Test score for  model is {test_score_dt}")

y_pred = dt_reg.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('MSE:',mse)
print('RMSE:',mse**(1/2))
```
