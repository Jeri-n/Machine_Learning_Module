```python
# importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore')

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_iris

```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['Species'] = iris.target
df.head()
```

```python
X = df.drop('Species',axis=1)
y = df.Species
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=17)

```

```python
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

print("The dimension of X_train is {}".format(X_train.shape))
print("The dimension of X_test is {}".format(X_test.shape))
```

```python
train_score_dt = dt_classifier.score(X_train, y_train)
test_score_dt = dt_classifier.score(X_test, y_test)

print(f"The Train score for  model is {train_score_dt}")
print(f"The Test score for  model is {test_score_dt}")

y_pred = dt_classifier.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print('MSE:',mse)
print('RMSE:',mse**(1/2))
```

```python
# trying with defaults 

dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)


train_score_dt = dt_classifier.score(X_train, y_train)
test_score_dt = dt_classifier.score(X_test, y_test)

print(f"The Train score for  model is {train_score_dt}")
print(f"The Test score for  model is {test_score_dt}")

y_pred = dt_classifier.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('MSE:',mse)
print('RMSE:',mse**(1/2))
```

```python
# trying with tuning

dt_classifier = DecisionTreeClassifier(criterion='entropy',max_depth=3)
dt_classifier.fit(X_train, y_train)


train_score_dt = dt_classifier.score(X_train, y_train)
test_score_dt = dt_classifier.score(X_test, y_test)

print(f"The Train score for  model is {train_score_dt}")
print(f"The Test score for  model is {test_score_dt}")

y_pred = dt_classifier.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('MSE:',mse)
print('RMSE:',mse**(1/2))
```
