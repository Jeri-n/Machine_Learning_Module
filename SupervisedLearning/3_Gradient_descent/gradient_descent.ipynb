```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

# y = 2x+3 -->function
```

```python
def gradient_descent(x,y):
    m_current = b_current = 0
    rate=0.08
    iter=10
    n= len(x)
    

    for i in range(iter):
        yp = m_current*x + b_current
        mse = (1/n)*sum((y-yp)**2)

        md = (2/n)*sum(x*(yp-y))
        m_current = m_current-rate*md
        bd = (2/n)*sum(yp-y)
        b_current = b_current-rate*bd
        
        print('m={} , b={} , iterartion={}, MSE = {}'.format(m_current, b_current,i,mse))

    
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
