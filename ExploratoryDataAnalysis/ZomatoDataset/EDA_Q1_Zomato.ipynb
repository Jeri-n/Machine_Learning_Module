```python
# 1. Loadf the country name from the dataset and merge with the main csv file

df_final=pd.merge(df,df_CountryCode,how="left")
df_final
```

```python
df_final.columns
```

##### Identify the Zomato's maximum transaction/records/observations based on country

```python
country_names = df_final['Country'].value_counts().index
```

```python
country_values = df_final['Country'].value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],radius=2,autopct='%1.2f%%')
plt.show
```

#### check most rated value

```python
ratings = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
plt.figure(figsize=(15,6))
sns.barplot(ratings['Aggregate rating'],ratings['Rating count'])
# sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings,hue='Rating color')
```

```python
# count plot

sns.countplot(ratings['Rating color'])
```
